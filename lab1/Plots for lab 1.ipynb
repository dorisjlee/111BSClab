### Oscilloscope

```python
f1 = [10.,31.6,100.,316.,1000.,3160.,10000.,31600,1e5,2e5,316000,3e5,1e6,3e6,5e6,6e6,7e6,5.5e6,7.5e6,8e6,8.5e6,9e6,9.5e6,1e7]
v1 = [480e-3,480e-3,480e-3,480e-3,480e-3,480e-3,480e-3,485e-3,485e-3,485e-3,490e-3,490e-3,490e-3,490e-3,495e-3,1.02,1.03,1.015,1.04,1.06,1.06,1.08,1.09,1.10]
```

```python
plt.title("Oscilloscope",fontsize=12)
plt.xlabel("Frequency (Hz)",fontsize=12)
plt.ylabel("$V_0$(V)",fontsize=12)
plt.plot(f1,v1,'o')
plt.plot(f1,v1,'r-')
```

##### On loglog scale

```python
plt.title("Oscilloscope",fontsize=12)
plt.xlabel("log Frequency (Hz)",fontsize=12)
plt.ylabel("log $V_0$(V)",fontsize=12)
plt.loglog(f1,v1,'o')
plt.loglog(f1,v1,'r-')
```
